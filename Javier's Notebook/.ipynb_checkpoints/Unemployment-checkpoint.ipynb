```python
pg_user = 'postgres'
pg_password = 'brassclaw1'
db_name = 'unemployment_db'

connection_string = f"{pg_user}:{pg_password}@localhost:5432/{db_name}"
engine = create_engine(f'postgresql://{connection_string}')
```
